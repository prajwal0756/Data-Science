In order to visualize multiple plots at one output we create different subplots as different axis 

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax1.plot(np.random.standard_normal(50).cumsum(), color="black", linestyle="dashed")
ax2.hist(np.random.standard_normal(50).cumsum(), color="orange", )
ax3.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30));
```

```python
fig, axes = plt.subplots(2, 3)
axes
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50, color="black", alpha=0.5)

fig.subplots_adjust(wspace=0, hspace=0)
```

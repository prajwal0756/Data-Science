To only replace empty values for one column, specify the column name for the DataFrame:

df.fillna({"discount": 130}, inplace=True)

There are many different ways to replace the nan value in dataset above we get some good examples but they cannot be used in all dataset as some required specific sensitive value so we use
Mean, Median and mode 

Mean = It is the average value od the whole data of the particular column 
Median= it is the middle value of the datas
mode = it is most frequently used data in whole dataset

```python
x = var["discount"].median()

var.fillna({"discount": x}, inplace=True)
var
```

```python
y= var["order_amount"].mean()

var.fillna({"order_amount": y}, inplace = True)
var
```

```python
z= var["city"].mode()[0]

var.fillna({"city":z}, inplace = True)
var
```

# Encoding Categorical Variables

### Two types od data:
    1. Nominal data(having no fix order)
    2. Ordinal data(Having fix order eg. marks grade)

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df= pd.read_csv('/Users/prajwalsubedi/Desktop/Data science/Data Sets/covid_dataset.csv',index_col=False)
```

```python
df= df.iloc[:,3:]
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test= train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Mild','Strong']])
```

```python
x_train[['Cough']]=oe.fit_transform(x_train[['Cough']])
x_test[['Cough']]=oe.transform(x_test[['Cough']])

# oe.fit_transform(x_train['Cough'])
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train=le.transform(y_train)
y_test=le.transform(y_test)
```

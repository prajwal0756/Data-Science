# Rename func for col

```python
def rename(name):
    if name=="Female":
        return "F"
    elif name=="Male":
        return "M"
    else:
        return name
        
```

```python

rename("Male")
```

```python
rename("Female")
```

```python
pd.read_csv('/Users/prajwalsubedi/Desktop/Data science/Data Sets/Telco_customer_churn.csv',converters={'Gender':rename}).shape
```

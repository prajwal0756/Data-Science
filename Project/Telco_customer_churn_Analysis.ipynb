```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, MinMaxScaler

preprocessor = ColumnTransformer([

    # One-hot
    
    ('ohe_internet_payment', OneHotEncoder(handle_unknown='ignore'), ['Internet Service','Payment Method']),
    ('ohe_city', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [0]),

    # Ordinal
    ('oe_gender', OrdinalEncoder(), ['Gender']),
    ('oe_partner', OrdinalEncoder(), ['Partner']),
    ('oe_contract', OrdinalEncoder(), ['Contract']),
    ('oe_senior_citizen', OrdinalEncoder(), ['Senior Citizen']),
    ('oe_dependents', OrdinalEncoder(), ['Dependents','Phone Service']),

    # Scaling
    ('scale', MinMaxScaler(), ['Tenure Months','Monthly Charges','Total Charges'])

], remainder='passthrough')
```

```python
from sklearn.feature_selection import SelectKBest, chi2
trf5= SelectKBest(score_func=chi2,k=8)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
# from sklearn.tree import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

trf6b= RandomForestClassifier()
```

```python
from sklearn.pipeline import Pipeline,make_pipeline
pipe = Pipeline([
    ('preprocessor',preprocessor),
    
    ('trf6b',trf6b)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
x_train
```

```python
pipe.named_steps['preprocessor'].transform(x_train)
```

```python
pipe.named_steps['preprocessor'].get_feature_names_out()
```

```python
y_pred =pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,x_train,y_train,cv=5, scoring='accuracy').mean()
```

```python
accuracy_score = accuracy_score(y_test,y_pred)
accuracy_score
```
